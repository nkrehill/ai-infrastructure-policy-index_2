```python
# Create a mock dataset matching real AESO data center connection queues
import pandas as pd

data = {
    'Project_ID': ['P201', 'P202', 'P203', 'P204', 'P205', 'P206'],
    'Project_Type': ['Data Centre', 'Data Centre', 'Wind Farm', 'Data Centre', 'Solar Array', 'Data Centre'],
    'AESO_Planning_Region': ['South', 'Central', 'South', 'Calgary', 'North', 'South'],
    'Requested_Load_MW': [800, 400, 150, 150, 200, 600],
    'Current_Status': ['Active', 'Under Review', 'Approved', 'Active', 'Approved', 'Under Review']
}

# Turn this into a Pandas DataFrame and save it as a local file
df = pd.DataFrame(data)
df.to_csv('aeso_queue_data.csv', index=False)
print("Success! 'aeso_queue_data.csv' has been created in your folder.")
```

```python
# Load the data file back into Pandas
queue_df = pd.read_csv('aeso_queue_data.csv')

# 1. Filter the dataset to ONLY show 'Data Centre' projects
data_centers = queue_df[queue_df['Project_Type'] == 'Data Centre']

# 2. Group the data centers by region and calculate the total requested Megawatts (MW)
regional_totals = data_centers.groupby('AESO_Planning_Region')['Requested_Load_MW'].sum().reset_index()

# 3. Display our findings
print("--- Total Requested Data Center Load by Alberta Region ---")
print(regional_totals)

# 4. Calculate total cumulative data center demand against Alberta's 1,200 MW interim cap
total_demand = data_centers['Requested_Load_MW'].sum()
print(f"\nTotal System-Wide Data Center Demand: {total_demand} MW")
print(f"Exceeds interim cap of 1,200 MW by: {total_demand - 1,200} MW")
```

```python
# Create a local text file simulating recent Alberta data center policy shifts
policy_text = """
GOVERNMENT OF ALBERTA - BILL 8 & BILL 12 AMENDMENTS
ELECTRICITY UTILITIES (DATA CENTRE INTEGRATION) ACT 

REGULATORY FRAMEWORK SUMMARY:
1. Effective immediately, a strict interim connection cap of 1,200 MW is placed on all new Large Load Infrastructure projects exceeding 50 MW.
2. All data center operators requesting grid access must undergo a mandatory Phase 2 Cluster Assessment by the AESO to evaluate localized grid stability.
3. Effective December 31, 2026, a new computing equipment levy will apply to all non-power-self-sufficient operations, adding operational cost friction.
4. Projects that demonstrate "behind-the-fence" self-generation utilizing carbon-neutral or co-located off-peak assets will receive fast-tracked regulatory approvals.
"""

with open('alberta_data_centre_policy.txt', 'w', encoding='utf-8') as f:
    f.write(policy_text.strip())

print("Created 'alberta_data_centre_policy.txt' successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Simulate the combined dataset (Pandas Data + Claude API qualitative scores)
combined_data = {
    'AESO_Planning_Region': ['South', 'Central', 'Calgary', 'Edmonton', 'North'],
    'Grid_Stress_Ratio': [18.2, 5.1, 1.2, 0.8, 0.4], # Calculated via Pandas in Phase 2
    'Regulatory_Risk_Score': [5, 4, 3, 2, 1]        # Evaluated via Claude in Phase 3
}

index_df = pd.DataFrame(combined_data)

# 2. Calculate the final "Infrastructure-Policy Friction Index Score"
# We multiply the two factors to create a combined hazard score
index_df['Friction_Index_Score'] = index_df['Grid_Stress_Ratio'] * index_df['Regulatory_Risk_Score']

print("--- Final Consolidated Friction Index Dataframe ---")
print(index_df)

# 3. Create a publication-quality visualization
plt.figure(figsize=(10, 5))
sns.set_theme(style="whitegrid")

# Create a horizontal bar chart
ax = sns.barplot(
    x="Friction_Index_Score", 
    y="AESO_Planning_Region", 
    data=index_df.sort_values(by="Friction_Index_Score", ascending=False),
    palette="flare"
)

# Customize look for a clean, minimalist professional aesthetic
plt.title('Alberta AI Infrastructure-Policy Friction Index by Region', fontsize=14, pad=15)
plt.xlabel('Composite Friction Score (Grid Stress × Regulatory Risk)', fontsize=12)
plt.ylabel('AESO Planning Region', fontsize=12)
sns.despine(left=True, bottom=True)

# Save the visualization directly to your project folder so we can use it in the README
plt.tight_layout()
plt.savefig('infrastructure_friction_index.png', dpi=300)
print("\nSuccess! 'infrastructure_friction_index.png' has been saved to your folder.")
```
